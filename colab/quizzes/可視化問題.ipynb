## 2問目

```python
df_sc012 = pd.read_csv('/Users/hironeko1234/Documents/PlayGround/Basic-Phython/Basic-Phython/colab/quizzes/matplotlib_quiz/scatter_data.csv')


c0 = df_sc012[df_sc012["class"]==0]
c1 = df_sc012[df_sc012["class"]==1]
c2 = df_sc012[df_sc012["class"]==2]

plt.scatter(c0["length (cm)"], c0["width (cm)"], color = "red", label = "class0")
plt.scatter(c1["length (cm)"], c1["width (cm)"], color = "blue", label = "class1")
plt.scatter(c2["length (cm)"], c2["width (cm)"], color = "green", label = "class2")
plt.xlabel("length(cm)")
plt.ylabel("width(cm)")
plt.title("クラス0~2のlengthとwidthに関する散布図")
plt.grid(True)
plt.legend()

```

```python
df_plt3 = pd.read_csv('/Users/hironeko1234/Documents/PlayGround/Basic-Phython/Basic-Phython/colab/quizzes/matplotlib_quiz/line_data.csv')
df_plt3

# 1つ目
xel = df_plt3["期間"]
y1el = df_plt3["件数"]
y2el = df_plt3["費用"]

plt.plot(xel, y1el, marker = "o")
plt.xlabel("期間")
plt.ylabel("件数")
plt.title("期間ごとの件数の推移")
plt.grid(True)

# x軸のラベルが重なるので45°で回転
plt.xticks(rotation=45)

# 2つ目
plt.bar(xel, y2el, width = 0.8)
plt.xlabel("期間")
plt.ylabel("費用(円)")
plt.title("期間ごとの費用の推移")
plt.grid(True)
```
